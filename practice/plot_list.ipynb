#### 前処理したベクトルをプロットしてみる。

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
csv_path = "../CSV/dataset1CSV/ascii/tlsh_ascii_4split_2label.csv"
df = pd.read_csv(csv_path, index_col=0)
```

```python
df.shape
```

```python
Backdoor_vector = df.iloc[0, :-1]
Packed_vector = df.iloc[237, :-1]
Ransom_vector = df.iloc[472, :-1]
```

```python
Backdoor_vector_numpy = Backdoor_vector.to_numpy()
Packed_vector_numpy = Packed_vector.to_numpy()
Ransom_vector_numpy = Ransom_vector.to_numpy()
print(type(Backdoor_vector_numpy))
```
