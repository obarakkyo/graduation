```python
with open(file1, "r") as f:
    f_json = json.load(f)
    max_num = 200
    api_lists = ["0" for i in range(max_num)] #100個の配列準備
    idx = 0 #インデックス初期化
    
    # thread_num = len(f_json["behavior"]["apistats"])

    for _, thread_id in enumerate(f_json["behavior"]["apistats"]):
        api_json = f_json["behavior"]["apistats"][thread_id]
        # print(api_json)
        for _, api_name in enumerate(api_json):
            for i in range(api_json[api_name]):
                if idx == max_num:
                    break
                api_lists[idx] = api_name
                idx += 1
                print(idx, api_lists[idx-1])
    
    print(api_lists)
            
    
```
